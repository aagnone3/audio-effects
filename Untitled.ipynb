```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from librosa.core import load, stft
```

```python
SR=16000
```

```python
x, fs = load('tone.wav', sr=SR)
x2, fs = load('rtone.wav', sr=SR)
print x.shape, x2.shape
```

```python
length = int(10.0 * SR / 2000)
print length
plt.plot(x[:length], 'b')
plt.plot(x2[:length], 'r')
```

```python
from librosa.display import specshow
spec1 = stft(x, n_fft=1024, hop_length=int(0.016*SR))
spec2 = stft(x2, n_fft=1024, hop_length=int(0.016*SR))
```

```python
specshow(spec1, sr=SR)
```

```python
specshow(spec2, sr=SR)
```
